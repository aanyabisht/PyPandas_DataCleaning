```python
import pandas as pd

df=pd.read_csv('Large_dataset.csv') #reading the csv file
```

```python
#fixing the unsorted column and initializing a new column to it using reset_index
```

```python
#Fix CustomerID first
clean_data=df
```

```python
clean_data=df.reset_index(drop=True)
clean_data['CustomerID']=clean_data.index +1 #index start from 1 
```

```python
#Standardizing text values
```

```python
#Fix Name
clean_data['Name']=clean_data['Name'].str.strip() #remove extra space
clean_data['Name']=clean_data['Name'].str.title() #make the first word capital
clean_data
```

```python
#Cleaning the age column 
```

```python
clean_data['Age']=clean_data['Age'].abs()
[clean_data['Age']<18] #LESS 18 IS 157
[clean_data['Age']>18] #GREATER THAN 18 IS 67
#75 null value and invalid age 157
```

```python
clean_data.loc[clean_data['Age']<18 ,'Age']=None #those who are less then 18 are not valid so converting them to Nan
avg=clean_data['Age'].mean() #getting the average age
avg#avg is 39
```

```python
clean_data['Age']=clean_data['Age'].fillna(avg) #replacing Nan with avg age
clean_data
```

```python
clean_data
```

```python
clean_data['Age']=clean_data['Age'].astype(int)
clean_data
```

```python
#fixing salary
clean_data['Salary'].mode() 
clean_data['Salary']=pd.to_numeric(clean_data['Salary'], errors='coerce') #converting the type to int/float and non numerice to Nan
```

```python

clean_data['Salary']=clean_data['Salary'].abs()
```

```python
avg_sal=clean_data['Salary'].mean() #avg salary 38273
clean_data

```

```python
clean_data['Salary']=clean_data['Salary'].fillna(avg_sal)
clean_data
```

```python
clean_data['Salary']=clean_data['Salary'].astype(int)

clean_data
```

```python
#using Quantile to see outliers in the dataset
```

```python
#quantile to see outiers in salary
q1=clean_data['Salary'].quantile(0.45)
q3=clean_data['Salary'].quantile(0.95)
iqr= q3-q1 #tell the total spread of the middle values in the dataset
lower=q1-1.5 *iqr
upper =q3+1.5*iqr
clean_data[(clean_data['Salary']>=lower) & (clean_data['Salary']<=upper)] #thats means our data have no outlier in salary
```

```python
clean_data['City'].info() #49 null-values type is correct
```

```python
clean_data['City']=clean_data['City'].str.strip()
clean_data['City']=clean_data['City'].str.title()
clean_data
```

```python
clean_data['City'].isnull().sum() #49
clean_data['City']=clean_data['City'].fillna("Unknown")
clean_data
```

```python
clean_data['Email'].info()
clean_data['Email'].str.strip()
clean_data['Email'].str.lower()
```

```python
#Cleaning dates
```

```python
clean_data['JoiningDate'].str.strip()
```

```python
print(clean_data['JoiningDate'].unique()) # getting unique dates
```

```python

clean_data['JoiningDate'] = clean_data['JoiningDate'].str.replace(
    'March 10 2023', '03-10-2023'
)
```

```python
clean_data['JoiningDate'] = clean_data['JoiningDate'].str.replace(
 '15/02/2023', '15-02-2023'
)
```

```python
clean_data['JoiningDate'] = clean_data['JoiningDate'].str.replace(
 '2023-01-15', '15-01-2023'
)
```

```python
clean_data['JoiningDate'] = clean_data['JoiningDate'].str.replace(
'2023/04/20' , '20-04-2023'
)
```

```python
clean_data['JoiningDate'].unique()
```

```python
clean_data['JoiningDate']=pd.to_datetime(clean_data['JoiningDate'], errors='coerce',dayfirst=False) # changing the type of date
```

```python
clean_data
```

```python
clean_data['JoiningDate'].isnull().sum() # getting the sum of null dates
```

```python
#Using mode get the most frequent date and replace the null
```

```python
clean_data['JoiningDate'].mode() #2023-04-20 
```

```python
clean_data['JoiningDate'] = clean_data['JoiningDate'].fillna(pd.Timestamp('2023-04-20')) #replacing the null
```
